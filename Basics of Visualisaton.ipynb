### Part 1: Data Types and Formats

1.1 Import necessary libraries:
- Pandas
- Numpy

```python
import pandas as pd
import numpy as np
```

### 1.2 Create a simple dataset:
- Name    | Age | Salary
- Alice   | 25  | 50000
- Bob     | 30  | 60000
- Charlie | 35  | 75000
- David   | 22  | 45000

```python
d = {'Name':['Alice','Bob','Charlie','David'],'Age':[25,30,35,22],'Salary':[50000,60000,75000,45000]}
df = pd.DataFrame(d)
```

1.3 Display the dataset:


```python
df
```

1.4 Check data types of columns:


```python
df.dtypes
```

1.5 Convert 'Age' to float:

```python
df['Age'] = df['Age'].astype(float)
df.dtypes
```

1.6 Save the dataset to CSV:


```python
df.to_csv('lab2.csv')
```

## Part 2: Data Visualization
2.1 Import Matplotlib library:

```python
import matplotlib.pyplot as plt
```

2.2 Plot a simple line chart of Name and Age 

```python
plt.plot(df['Name'], df['Age'],color='red')  
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('LINE CHART')
plt.show()  
```

2.3 Create a bar chart for Salary:


```python
bar = plt.bar(df['Name'],df['Salary'],color='purple')
plt.xlabel('Index')
plt.ylabel('Salary')
plt.title('BAR CHART')
plt.show()
```

2.4 Scatter plot for Age and Salary:


```python
plt.scatter(df['Age'],df['Salary'],color='green')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('SCATTER PLOT')
plt.show()
```

2.5 Save the visualizations:


```python
bar = plt.bar(df['Name'],df['Salary'],color='purple')
plt.savefig('bar.jpg')
```
